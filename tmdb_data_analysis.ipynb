```python
df.drop("Unnamed: 0",axis=1,inplace=True)
df
```

```python
df["title"] = df["title"].astype("object")
df["original_title"] = df["original_title"].astype("object")
df["overview"] = df["overview"].astype("object")
```

```python
df["original_language"] = df["original_language"].astype("category")
df["release_year"] = df["release_year"].astype("category")
```

```python
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")
```

```python
df["vote_average"] = pd.to_numeric(df["vote_average"].astype("float32"), errors="coerce")
df["vote_count"] = pd.to_numeric(df["vote_count"].astype("int32"), errors="coerce")
df["popularity"] = pd.to_numeric(df["popularity"].astype("float32"), errors="coerce")
df["overview_word_count"] = pd.to_numeric(df["overview_word_count"].astype("int32"), errors="coerce")

df.dtypes
```

```python
import ast
language_table = df["original_language"].value_counts().reset_index()
language_table.columns = ["Language", "Count"]
print(language_table)
```

```python
import ast
def convert_genres(x):
    if isinstance(x, str):  
        return ast.literal_eval(x)
    return x  

df["genres"] = df["genres"].apply(convert_genres)

def get_genre_names(x):
    if isinstance(x, list) and len(x) > 0 and isinstance(x[0], dict):
        return [g["name"] for g in x]
    return x

df["genres"] = df["genres"].apply(get_genre_names)


genre_table = df["genres"].explode().value_counts().reset_index()
genre_table.columns = ["Genre", "Count"]
print(genre_table)

print(df["genres"].value_counts())
print(df["genres"].value_counts().count())
```
