```python
import requests
import pandas as pd

all_dfs = [] 

for i in range(1, 150):
    response = requests.get(
        f'https://api.themoviedb.org/3/movie/top_rated?api_key=ced2fc8ea193b30272c2e6e516ebebee&language=en-US&page={i}'
    )
    data = response.json().get('results', [])
    if data:  
        temp_df = pd.DataFrame(data)[[
             'title', 'release_date', 'vote_average','original_title','genre_ids',
            'vote_count', 'popularity', 'original_language', 'overview'
        ]]
        all_dfs.append(temp_df)


df = pd.concat(all_dfs, ignore_index=True)

```

```python
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")
df["release_year"] = df["release_date"].dt.year
df
```

```python
df["genre_ids"].head().apply(type)
```

```python
genre_map = {
    28: "Action",
    12: "Adventure",
    16: "Animation",
    35: "Comedy",
    80: "Crime",
    99: "Documentary",
    18: "Drama",                            ##referred from tmdb api site
    10751: "Family",
    14: "Fantasy",
    36: "History",
    27: "Horror",
    10402: "Music",
    9648: "Mystery",
    10749: "Romance",
    878: "Science Fiction",
    10770: "TV Movie",
    53: "Thriller",
    10752: "War",
    37: "Western"
}
```

```python
df["genres"] = df["genre_ids"].apply(
    lambda lst: [genre_map.get(g, "Unknown") for g in lst]
)
df
```

```python
df["vote_count"] = pd.to_numeric(df["vote_count"], errors="coerce")
df["popularity"] = pd.to_numeric(df["popularity"], errors="coerce")
```

```python
df["vote_count"] = df["vote_count"].fillna(0)
df["popularity"] = df["popularity"].fillna(df["popularity"].median())
```

```python
df = df[df["vote_count"] >= 0]
df = df[df["popularity"] >= 0]
df
```

```python
language_map = {
    "en": "English",
    "fr": "French",
    "ja": "Japanese",
    "it": "Italian",
    "es": "Spanish",
    "ko": "Korean",
    "de": "German",
    "zh": "Chinese",
    "cn": "Chinese",
    "ru": "Russian",
    "pt": "Portuguese",
    "hi": "Hindi",
    "sv": "Swedish",
    "da": "Danish",                         ##referred from tmdb api site
    "tr": "Turkish",
    "pl": "Polish",
    "fa": "Persian",
    "no": "Norwegian",
    "th": "Thai",
    "nl": "Dutch",
    "fi": "Finnish",
    "hu": "Hungarian",
    "sr": "Serbian",
    "cs": "Czech",
    "te": "Telugu",
    "ar": "Arabic",
    "uk": "Ukrainian",
    "xx": "Unknown",
    "id": "Indonesian",
    "bs": "Bosnian",
    "bn": "Bengali",
    "el": "Greek",
    "ga": "Irish",
    "is": "Icelandic",
    "hy": "Armenian",
    "et": "Estonian",
    "ml": "Malayalam",
    "ro": "Romanian",
    "la": "Latin",
    "lv": "Latvian",
    "sh": "Serbo-Croatian"
}

df["original_language"] = df["original_language"].map(language_map).fillna(df["original_language"])
df
```

```python
df['original_language'].value_counts()
```

```python
df[df.duplicated(subset=["original_title", "release_year"], keep=False)]
df.drop_duplicates(subset=["original_title", "release_year"], inplace=True)
df
```

```python
df["title"] = df["title"].astype(str)
df["title"] = df["title"].str.strip()
df["title"] = df["title"].str.replace(r"\s+", " ", regex=True)
df["title"] = df["title"].str.title()
df["title"] = df["title"].replace("", None)
df["title"] = df["title"].fillna("Unknown Title")
df
```

```python
df["overview"] = df["overview"].astype(str)
df["overview"] = df["overview"].str.strip()
df["overview"] = df["overview"].str.replace(r"\s+", " ", regex=True)
df["overview"] = df["overview"].replace("nan", None)
df["overview"] = df["overview"].fillna("No overview available")

df["overview_word_count"] = df["overview"].apply(lambda x: len(x.split()))
df
```

```python
df.columns
```

```python
new_order = [
    "title",
    "original_title",
    "overview",
    "overview_word_count",
    "release_date",
    "release_year",
    "original_language",
    "genres",
    "vote_average",
    "vote_count",
    "popularity"
]

df = df[new_order]
df
```

```python
df.info()
```

```python
df = df.drop(columns=["Unnamed: 0"], errors="ignore")

df["title"] = df["title"].astype("string")
df["original_title"] = df["original_title"].astype("string")
df["overview"] = df["overview"].astype("string")

df["original_language"] = df["original_language"].astype("category")
df["release_year"] = df["release_year"].astype("category")

df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")

df["vote_average"] = pd.to_numeric(df["vote_average"], errors="coerce")
df["vote_count"] = pd.to_numeric(df["vote_count"], errors="coerce")
df["popularity"] = pd.to_numeric(df["popularity"], errors="coerce")
df["overview_word_count"] = pd.to_numeric(df["overview_word_count"], errors="coerce")


df.dtypes
```
